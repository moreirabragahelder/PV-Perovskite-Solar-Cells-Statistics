# PV Perovskite Solar Cells Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import linregress  # Importing necessary libraries
```

# Power Conversion Efficiency

```python
# Load data from a CSV file into a pandas DataFrame
data = pd.read_csv('/home/helder-braga/Documentos/PSCs Resuls/IPS/Estabilização MAPI.csv')

# Create a DataFrame from the loaded data
df = pd.DataFrame(data)

# Set Seaborn style and color palette for better visualization
sns.set_style("ticks")  # Alternative styles: "darkgrid", "white", "whitegrid"
custom_palette = sns.color_palette("Blues", 8)  # Choosing a blue color palette
sns.set_palette(custom_palette)

# Create a new figure with specified size
plt.figure(figsize=(8, 6))

# Create a boxplot to display the distribution of PCE values across different concentrations
sns.boxplot(x="Concentration", y="PCE", data=df, whis=[0, 100], width=0.5)

# Overlay a strip plot on top of the boxplot to show individual data points
sns.stripplot(x="Concentration", y="PCE", data=df, size=8, color="darkred", linewidth=1, jitter=True)

# Set font sizes for tick labels
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Set titles and axis labels with appropriate font sizes
plt.title("MAPI", fontsize=20)
plt.ylabel("PCE (%)", fontsize=20)
plt.xlabel("Days", fontsize=20)

# Save the figure as a PNG file with a transparent background
plt.savefig('PCE MAPI.png', transparent=True)

# Display the plot
plt.show()
```

# PCE Degrades Over Time

```python
# Load the data from a CSV file into a pandas DataFrame
media_MAPI = pd.read_csv('/home/helder-braga/Documentos/PSCs Resuls/IPS/Media Estabilização MAPI.csv')

# Create a DataFrame from the loaded data
df1 = pd.DataFrame(media_MAPI)

# Set the Seaborn style for the plot
sns.set_style("ticks")  # Optional, styles can be "darkgrid", "whitegrid", "white", etc.

# Create a new figure with specified size
plt.figure(figsize=(8, 6))

# Plot scatter points for the PCE values against Concentration
sns.scatterplot(x='Concentration', y='PCE', data=df1, s=100, color='blue')  # Adjust 's' for point size

# Add error bars to represent the standard deviation of PCE
plt.errorbar(
    x=df1['Concentration'], 
    y=df1['PCE'], 
    yerr=df1['std_PCE'], 
    fmt='o', 
    capsize=5, 
    label='Error Bars', 
    color='black'
)

# Add a linear regression line without scatter points
sns.regplot(
    x=df1['Concentration'], 
    y=df1['PCE'], 
    data=df1, 
    scatter=False, 
    label='Linear Fit', 
    color='red'
)

# Set font sizes for axis ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Customize the plot with a title and axis labels
plt.title('MAPI', fontsize=20)
plt.ylabel("Average PCE (%)", fontsize=20)
plt.xlabel("Days", fontsize=20)

# Compute the linear regression to determine the slope
slope, intercept, r_value, p_value, std_err = linregress(df1['Concentration'], df1['PCE'])

# The slope (coefficient of the regression line) represents the rate of PCE degradation
coeficiente_da_reta_PCEMAPI = slope * 100  # Convert to percentage

# Display the slope value on the plot
plt.text(
    0.7, 0.9, 
    f'Lose PCE: {coeficiente_da_reta_PCEMAPI:.2f}%', 
    transform=plt.gca().transAxes, 
    fontsize=12, 
    verticalalignment='top', 
    bbox=dict(boxstyle='round, pad=0.5', facecolor='white', alpha=0.5)
)

# Uncomment the following lines to display individual PCE values on the graph
# for i, row in df1.iterrows():
#     plt.text(row['Concentration'], row['PCE'] + 0.2, f'{row["PCE"]:.2f}', 
#              ha='center', va='bottom', fontsize=12, color='green')

# Save the plot as a PNG file with a transparent background
plt.savefig('MAPI PCE Average.png', transparent=True)

# Uncomment the next line to show a legend for the error bars and regression line
# plt.legend()

# Display the plot
plt.show()

# Print the slope value, which indicates the rate of PCE degradation
print("Linear fit slope coefficient:", coeficiente_da_reta_PCEMAPI)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Load the data from a CSV file into a pandas DataFrame
media_MAPI = pd.read_csv('/home/helder-braga/Documentos/PSCs Resuls/IPS/Media Estabilização MAPI.csv')

# Create a DataFrame from the loaded data
df2 = pd.DataFrame(media_MAPI)

# Set the Seaborn style for the plot
sns.set_style("ticks")  # Other options: "darkgrid", "whitegrid", "white", etc.

# Create a new figure with a specified size
plt.figure(figsize=(8, 6))

# Scatter plot: Visualizing individual data points
sns.scatterplot(x='Concentration', y='VOC', data=df2, s=100, color='blue', label='Data Points')

# Error bars: Showing standard deviation
plt.errorbar(x=df2['Concentration'], y=df2['VOC'], yerr=df2['std_VOC'], fmt='o', capsize=5, label='Error Bars', color='black')

# Linear regression: Analyzing the trend
sns.regplot(x=df2['Concentration'], y=df2['VOC'], data=df2, scatter=False, label='Linear Fit', color='red')

# Set font sizes for x and y ticks
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Set the title and axis labels
plt.title('MAPI', fontsize=20)
plt.ylabel("Average VOC (V)", fontsize=20)
plt.xlabel("Days", fontsize=20)

# Perform linear regression to determine the rate of VOC degradation
slope, intercept, r_value, p_value, std_err = linregress(df2['Concentration'], df2['VOC'])

# The slope represents the VOC degradation rate (expressed as a percentage)
coeficiente_da_reta_VOCMAPI = slope * 100

# Display the degradation rate on the plot
plt.text(0.7, 0.9, f'VOC Loss: {coeficiente_da_reta_VOCMAPI:.2f}%', 
         transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', 
         bbox=dict(boxstyle='round, pad=0.5', facecolor='white', alpha=0.5))

# Save the figure as a PNG file with a transparent background
plt.savefig('MAPI_VOC_Average.png', transparent=True)

# Show the plot
plt.legend()
plt.show()

# Print the calculated slope value
print("Linear fit slope coefficient:", coeficiente_da_reta_VOCMAPI)
```

# JSC Changes Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from a CSV file containing JSC stability results for MAPI perovskite solar cells
data = pd.read_csv('/home/helder-braga/Documentos/PSCs Resuls/IPS/Estabilização MAPI.csv')

# Create a DataFrame from the CSV data
df = pd.DataFrame(data)

# Set Seaborn style and color palette for better visualization aesthetics
sns.set_style("ticks")  # Choosing a minimalistic style without grid lines
custom_palette = sns.color_palette("Blues", 8)  # Using a blue gradient color palette
sns.set_palette(custom_palette)  # Apply the custom palette to Seaborn plots

# Create a new figure with specified size
plt.figure(figsize=(8, 6))

# Generate a boxplot to visualize the distribution of JSC values for different concentrations
sns.boxplot(x="Concentration", y="JSC", data=df, whis=[0, 100], width=0.5)

# Overlay a strip plot to show individual data points with slight jitter for better visibility
sns.stripplot(x="Concentration", y="JSC", data=df, size=8, color="darkred", linewidth=1, jitter=True)

# Adjust font sizes for better readability
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Set the plot title and axis labels with appropriate font sizes
plt.title("MAPI", fontsize=20)
plt.ylabel("JSC ($mA.cm^{-2}$)", fontsize=20)  # Corrected unit notation
plt.xlabel("Days", fontsize=20)

# Save the plot as a PNG image with a transparent background
plt.savefig('JSC_MAPI.png', transparent=True)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Load the dataset containing JSC stability results for MAPI perovskite solar cells
df3 = pd.read_csv('/home/helder-braga/Documentos/PSCs Resuls/IPS/Media Estabilização MAPI.csv')

# Set Seaborn style for better visualization
sns.set_theme(style="ticks")

# Create a figure with appropriate size
plt.figure(figsize=(8, 6))

# Scatter plot for visualizing individual data points
sns.scatterplot(x='Concentration', y='JSC', data=df3, s=100, color='blue', label="Data Points")

# Error bars to show standard deviation in JSC values
plt.errorbar(x=df3['Concentration'], y=df3['JSC'], yerr=df3['std_JSC'], fmt='o', capsize=5, color='black', label='Error Bars')

# Linear regression fit to observe trend
sns.regplot(x='Concentration', y='JSC', data=df3, scatter=False, color='red', label='Linear Fit')

# Customize plot appearance
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("MAPI - JSC Stability", fontsize=20, fontweight="bold")
plt.ylabel(r"Average JSC ($\mathrm{mA \cdot cm^{-2}}$)", fontsize=18)
plt.xlabel("Days", fontsize=18)

# Compute linear regression to determine degradation rate
slope, intercept, r_value, p_value, std_err = linregress(df3['Concentration'], df3['JSC'])

# Convert slope to percentage degradation
JSC_degradation_rate = slope * 100

# Display degradation rate on the plot
plt.text(0.7, 0.9, f'JSC Loss: {JSC_degradation_rate:.2f}%', transform=plt.gca().transAxes, 
         fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.5))

# Save the figure with high resolution
plt.savefig("MAPI_JSC_Average.png", dpi=300, transparent=True, bbox_inches="tight")

# Show the final plot
plt.legend()
plt.show()

# Print the calculated linear regression slope
print("Linear regression slope (JSC degradation rate):", JSC_degradation_rate)
```
